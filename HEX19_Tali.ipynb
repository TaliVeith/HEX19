```python
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
import matplotlib.pyplot as plt

def draw_hexagons():
    fig, ax = plt.subplots(nrows=1, ncols=1)
    fig.set_size_inches((15,10))
    
    patches = []
    probabilities = []
    for j,row in enumerate(Probabilities_after_n_steps_on_board):
        for i,element in enumerate(row):
            if (Board[j,i]):
                points = np.array([[i-0.3,j-0.7],[i+0.3,j-0.7],[i+0.6,j],[i+0.3,j+0.7],[i-0.3,j+0.7],[i-0.6,j]])
                polygon = Polygon(points, True)
                patches.append(polygon)
                probabilities.append(Probabilities_after_n_steps_on_board[j,i])

    p = PatchCollection(patches, alpha=0.4)
    p.set_array(np.array(probabilities))
    ax.add_collection(p)
    plt.colorbar(p, ax=ax)
    
    for j,row in enumerate(Probabilities_after_n_steps_on_board):
        for i,element in enumerate(row):
            if (Board[j,i]):
                plt.text(i-0.2,j+0.1,'%.2f' % (Probabilities_after_n_steps_on_board[j,i]*100))
    
    plt.inferno()
    plt.xlim(-1, Probabilities_after_n_steps_on_board.shape[1])
    plt.ylim(Probabilities_after_n_steps_on_board.shape[0], -1)
    ax.set_title("Walked " + str(number_of_steps) + " steps.")
    plt.show()
    return fig, ax
```
